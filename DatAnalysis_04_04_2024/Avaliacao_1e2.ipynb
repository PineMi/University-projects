<strong>c) Usando funções programadas pelo estudante (Com exceção das visualizações

```python
# Função para calcular a média
def calc_media(data):
    return sum(data) / len(data)

# Função para calcular a mediana
def calc_mediana(data):
    n = len(sorted(data))
    if n % 2 == 0:
        return (sorted(data)[n//2 - 1] + sorted(data)[n//2]) / 2
    else:
        return sorted(data)[n//2]

# Função para calcular a variância populacional
def calc_vpop(data):
    return sum((x - calc_media(data)) ** 2 for x in data) / len(data)

# Função para calcular a variância amostral
def calc_vamo(data):
    return sum((x - calc_media(data)) ** 2 for x in data) / (len(data) - 1)

# Função para calcular o desvio padrão populacional
def calc_dp_pop(data):
    return calc_vpop(data) ** 0.5

# Função para calcular o desvio padrão amostral
def calc_dp_amo(data):
    return calc_vamo(data) ** 0.5

# Função para calcular o coeficiente de variação populacional
def calc_coef_vpop(data):
    mean = calc_media(data)
    return (calc_dp_pop(data) / mean) * 100

# Função para calcular o coeficiente de variação amostral
def calc_coef_vamo(data):
    mean = calc_media(data)
    return (calc_dp_amo(data) / mean) * 100

# Função para calcular os quartis
def calc_quartis(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    q1 = sorted_data[int(n * 0.25)]
    q2 = calc_mediana(data)
    q3 = sorted_data[int(n * 0.75)]
    return q1, q2, q3

quartis = calc_quartis(conjunto)
```

```python
print(f'''
Média: {calc_media(conjunto)}
Mediana: {calc_mediana(conjunto)}
Variância Populacional: {calc_vpop(conjunto)}
Variância Amostral: {calc_vamo(conjunto)}
Desvio Padrão Populacional: {calc_dp_pop(conjunto)}
Desvio Padrão Amostral: {calc_dp_amo(conjunto)}
Coeficiente de Variação Populacional: {calc_coef_vpop(conjunto)}%
Coeficiente de Variação Amostral: {calc_coef_vamo(conjunto)}%
Quartis: Q1={quartis[0]}, Q2={quartis[1]}, Q3={quartis[2]}
''')
```

<strong>1) Aplicar a regressão linear em uma base de dados real (base quantitativa).</strong>

#### $r = \frac{n\sum{xy}-\sum{x}\sum{y}}{\sqrt{[n\sum{x^2}-(\sum{x})^2][n\sum{y^2}-(\sum{y})^2]}}$

### $a = r  \frac{\sigma_y}{\sigma_x} = \frac{n\sum{xy}-\sum{x}\sum{y}}{n\sum{x^2}-(\sum{x})^2}$

### $b = \mu_y - a\mu_x = \frac{\sum{y}}{n}a\frac{\sum{x}}{n}$

```python
dados = planets_data[["mass", "orbital_period"]].dropna()
dados.head()
```

<strong>a) Encontrar os coeficiêntes a e b, usando as fórmulas dadas em aula. Fazer os gráficos para
a base de dados e para a regressão linear.</strong>

Versão manuscrita enviada em PDF

```python
n = len(dados)
Ex = sum(dados["mass"])
Ey = sum(dados["orbital_period"]) 
Exy = sum([x*y for x,y in zip(dados["mass"], dados["orbital_period"])])
Ex2 = sum([x*x for x in dados["mass"]])
Ey2 =  sum([y*y for y in dados["orbital_period"]])
Ex_2, Ey_2 = Ex**2, Ey**2
```

```python
r = (n*Exy-Ex*Ey)/((n*Ex2-Ex_2)*(n*Ey2-Ey_2))**0.5
a = (n*Exy-Ex*Ey)/(n*Ex2-Ex_2)
b = Ey/n-a*Ex/n
a,b
```

```python
plt.scatter(dados["mass"],dados["orbital_period"])
plt.axline((0, b), slope=a, linewidth=2, color='r')
plt.show()
```
